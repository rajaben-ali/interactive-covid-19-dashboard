```
# All packages to import here
import streamlit as st
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import zscore
import matplotlib.pyplot as plt
import plotly.express as px
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
```

```
# Load the csv files
df_confirmed = pd.read_csv("./data/time_series_covid19_confirmed_global.csv")
df_deaths = pd.read_csv("./data/time_series_covid19_deaths_global.csv")
df_recovered = pd.read_csv("./data/time_series_covid19_recovered_global.csv")
df_population = pd.read_csv("./data/population_by_country.csv")
```

```
# Data Preparation - Cumulative data
# Confirmed cases
df_confirmed_cumul = df_confirmed.melt(id_vars=["Province/State", "Country/Region",'Lat', "Long"], var_name='Date', value_name='Count')
# Death cases
df_deaths_cumul = df_deaths.melt(id_vars=["Province/State", "Country/Region",'Lat', "Long"], var_name='Date', value_name='Count')
# Recovered cases
df_recovered_cumul = df_recovered.melt(id_vars=["Province/State", "Country/Region",'Lat', "Long"], var_name='Date', value_name='Count')
```

```
confirmed_n = pd.merge(df_confirmed_cumul, df_population,how = 'inner', left_on = 'Country/Region' , right_on = 'Country')
confirmed_n['Population_million'] = confirmed_n['Population']/1000000
confirmed_n['number_per_capita']= confirmed_n['Count']/confirmed_n['Population_million']
confirmed_n = confirmed_n.round({'number_per_capita': 0})
confirmed_n["Date"] = confirmed_n["Date"].astype('datetime64[ns]')
```

```
def normalized(df_type):
    df = pd.merge(df_type, df_population,how = 'inner', left_on = 'Country/Region' , right_on = 'Country')
    df['Population_million'] = df['Population']/1000000
    df['number_per_capita']= df['Count']/df['Population_million']
    df = df.round({'number_per_capita': 0})
    df["Date"] = df["Date"].astype('datetime64[ns]')
    return df
    
```

```
def aggregation_for_country(df, s_csv_country):
    df = df.groupby(["Date", 'Country/Region'], as_index=False).agg({'number_per_capita':'sum','Count':'sum'}).set_index('Date')
    df = df[df['Country/Region']==s_csv_country]
    return df
```

```
# Prepare Data
options_csv_country = np.unique(np.concatenate((df_confirmed_cumul["Country/Region"].unique(),
  df_deaths_cumul["Country/Region"].unique(),
  df_recovered_cumul["Country/Region"].unique()), 0))

# This will be replaced by the multiselect widget - to check if the plot works, update the following variable
s_csv_country = 'France' # Only one country for this plot

# TODO: plot the selected country's death line chart, confirmed chart and recovered line chart in one
s_csv_country
```

```

confirmed = normalized(df_confirmed_cumul)
confirmed = aggregation_for_country(confirmed, s_csv_country=)
recovered = normalized(df_recovered_cumul)
recovered = aggregation_for_country(recovered,s_csv_country=)
```

```
confirmed = confirmed.rename(columns={'number_per_capita':'confirmed_number'})
recovered = recovered.rename(columns={'number_per_capita':'recovered_number'})

df = pd.merge(confirmed,recovered, how='left',left_index = True, right_index = True).reset_index()
fig = px.scatter(df, x = 'Date', y = ['confirmed_number', 'recovered_number'])
fig.update_layout(
    title="The trend of Recovered and Confirmed Covid cases",
    xaxis_title="Date",
    yaxis_title="Number of cases"
       
    )
fig.show()
```

```
# import and open the csv files

confirmed = pd.read_csv("./data/time_series_covid19_confirmed_global.csv")
deaths = pd.read_csv("./data/time_series_covid19_deaths_global.csv")
recovered = pd.read_csv("./data/time_series_covid19_recovered_global.csv")

# drop the columns of lat and long

confirmed = confirmed.drop(columns=["Lat", "Long"])
deaths = deaths.drop(columns=['Lat', 'Long'])
recovered = recovered.drop(columns=['Lat', 'Long'])

# melt the data by date
dates_confirmed = confirmed.columns[2:]
confirmed_melted = confirmed.melt(id_vars=['Province/State', 'Country/Region'], value_vars=dates_confirmed, var_name='Date',value_name='Confirmed')
dates_deaths = deaths.columns[2:]
deaths_melted = deaths.melt(id_vars=['Province/State', 'Country/Region'], value_vars=dates_deaths, var_name='Date',value_name='Deaths')
dates_recovered = recovered.columns[2:]
recovered_melted = recovered.melt(id_vars=['Province/State', 'Country/Region'], value_vars=dates_recovered, var_name='Date',value_name='Recovered')

# convert the date time to standard time format

confirmed_melted["Date"] = confirmed_melted["Date"].apply(
    lambda x: datetime.datetime.strptime(x, '%m/%d/%y').date())

deaths_melted["Date"] = deaths_melted["Date"].apply(
    lambda x: datetime.datetime.strptime(x, '%m/%d/%y').date())

recovered_melted["Date"] = recovered_melted["Date"].apply(
    lambda x: datetime.datetime.strptime(x, '%m/%d/%y').date())
```
