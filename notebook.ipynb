# Deaths

```
import pandas as pd
```

```
deaths = pd.read_csv('data/time_series_covid19_deaths_global.csv')

df_deaths = pd.DataFrame(deaths)
df_deaths.head()
```

```
df_deaths_p=df_deaths.melt(id_vars = ["Province/State", "Country/Region",'Lat', "Long"], var_name = 'Date',value_name =  'Count')
df_deaths_p
```

```
df_deaths_p.sort_values(by = 'Count', axis = 0,ascending= False)
df_deaths_p
```

# Recovered
